```r
head(df)
```

```r
  keyvals <- ifelse(
    df$avg_logFC < -1 & df$p_val_adj < 0.05, 'orchid4',
      ifelse(df$avg_logFC > 1 & df$p_val_adj < 0.05, 'gold1',
        'gray'))
  keyvals[is.na(keyvals)] <- 'gray'
  names(keyvals)[keyvals == 'gold1'] <- 'EDAD_enriched'
  names(keyvals)[keyvals == 'gray'] <- 'not_significant'
  names(keyvals)[keyvals == 'orchid4'] <- 'ODAD_enriched'
```

```r
pdf(width=9,height=7,file="3A_volcano_spatial_gold1_orchid4_revision.pdf") #uncomment to save as pdf. Change width and height accordingly

options(repr.plot.height = 7, repr.plot.width = 9, dpi = 500)
EnhancedVolcano(df,
                lab = row.names(df),
                    selectLab = c('IFIT2','CCL2','ISG15','IFIT3','IL1B','1L17A','IL1RL1','SERPINE1','SPP1','COL1A1','COL1A2','COL12A1','MMP2','MMP9','POSTN','SMAD7','LUM','SMAD6'),
                x = 'avg_logFC',
                y = 'p_val_adj',
                title = 'ODAD_vs_EDAD_spatial_edgeR_deg',
                subtitle = NULL,
                caption = NULL,
                xlab = bquote(~avg ~log[2] ~FC),
                ylab = bquote(~-log[10] ~ p_val_adj),
                pCutoff = 0.05,
                FCcutoff = 1, # it means log2FC here!
                xlim = c(-6,6),
                ylim = c(0,23),
                pointSize = 2.5,
                labSize = 5,
                labFace = 'bold.italic',
                legendPosition = 'top',
                legendLabSize = 15,
                drawConnectors = T,
                widthConnectors = 0.5,
                boxedLabels = FALSE,
                colCustom = keyvals,
                colAlpha = 0.3) + coord_flip()

dev.off() 
```
