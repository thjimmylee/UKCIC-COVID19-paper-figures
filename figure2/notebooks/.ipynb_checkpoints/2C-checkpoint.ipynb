```python
import pandas as pd

import plotnine
from plotnine import *

dpi=300
```

```python
df=pd.read_csv('../data/Fig2C.csv')
```

```python
p1=(
    ggplot() 
    + geom_point(
        data=df, 
        mapping=aes(
            x='Donor_AltName',
            y='Pathology', 
            size='n_ROIs', 
            colour='Pathology'
        )
    )
    + geom_text(
        data=df,
        mapping=aes(
            x='Donor_AltName', 
            y='Pathology', 
            label='n_ROIs'), 
        size=30
    )   
    + theme_bw() 
    + theme(
        axis_text_x=element_text(angle=60, vjust=1, hjust=1, size=30), 
        axis_text_y=element_text(size=30),
        title=element_text(size=36), 
        axis_title=element_text(size=36), 
        panel_grid_major=element_blank(),
        figure_size=(28,6), 
        dpi=dpi,
        legend_position='none'
    )  
    + scale_size_continuous(range=[20,30]) 
    + scale_y_discrete(limits=['PRES', 'EDAD', 'MDAD', 'ODAD'][::-1]) 
    + scale_x_discrete(limits=df.Donor_AltName.unique()) 
    + scale_colour_manual(
        values=[
            '#DFA1A0',
            '#D17372',
            '#C14443',  
            '#ECD6D6' 
    ]) 
    + labs(x='Donors', y='Pathology', title='Number of ROIs collected')
)

# ggsave(p1, 'figures/fig2c.pdf', format='pdf', limitsize=False)
p1
```
